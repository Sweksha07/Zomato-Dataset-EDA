```python
dfc=pd.read_excel("Country-Code.xlsx")
```

```python
dfc
```

```python
df_final=pd.merge(df,dfc,on="Country Code",how="left")
```

```python
df_final.head()
```

```python
df_final.Country.value_counts()
```

```python
country_name=df_final.Country.value_counts().index
```

```python
country_name
```

```python
country_val=df_final.Country.value_counts().values
```

```python
#Pie Chart
```

```python
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
plt.show
```

Observation: We can See That India has the Largest Usage of Zomato and the next two countries are US and UK

```python
#Dealing with Numerical records
```

```python
df_final.columns
```

```python
ratings=df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

### Observation
     1.When the rating is between 4.5 to 4.9 -->Excellent
     2.When the rating is between 4.1 to 4.5-->Very Good
     3.When the rating is between 3.5 to 3.9-->Good
     4.When the rating is between 2.5 to 3.4--> Average
     5.When the rating is between 1.8 to 2.4-->Poor
     6.People who have'nt Rated 

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize']=(18,8)
sns.barplot(x="Aggregate rating",y="Rating count",hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

### Observation-->
    1.We can see that most of the people have'nt rated here
    2.Secondly average rating is between 2.5 to 3.9

```python
#countplot

sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
plt.show()
```
